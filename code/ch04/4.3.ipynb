# 数值微分

## 导数

```python
# 不好的实现示例
def numerical_diff(f, x):
    h = 10e-50
    return (f(x + h) - f(x)) / h
```

```python
def numerical_diff(f, x):
    h = 1e-4
    return (f(x + h) - f(x - h)) / (2 * h)
```

## 数值微分的例子

```python
def function_1(x):
    return 0.01 * x ** 2 + 0.1 * x
```

```python
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)
plt.show()
```

```python
print(numerical_diff(function_1, 5))
print(numerical_diff(function_1, 10))
```

## 偏导数

```python
def function_2(x):
    return x[0] ** 2 + x[1] ** 2
```

```python
def function_tmp1(x0):
    return x0 ** 2 + 4.0 ** 2

numerical_diff(function_tmp1, 3.0)
```
