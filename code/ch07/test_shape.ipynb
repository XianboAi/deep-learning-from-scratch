```python
x = np.ones((10, 3, 8, 8))
y = np.zeros((10, 3, 6, 6, 3, 3))
for i in range(6):
    for j in range(6):
        y[:, :, i, j, :, :] = x[:, :, i:i+3, j:j+3]
```
