```python
def softmax(a):
    sum_exp = np.sum(a, axis=1, keepdims=True)
    return a / sum_exp
```

```python
a = np.array([[1, 2],
              [10, 20]])
s = np.sum(a, axis=1)
print(a.shape)
print(s.shape)
print(a / s)
```
