```python
import numpy as np
```

# 恒等函数和softmax函数

```python
def softmax(a):
    y = np.exp(a) / np.sum(np.exp(a))
    return y
```

# 实现softmax函数时的注意事项

```python
# 溢出
a = np.array([1010, 1000, 900])
np.exp(a) / np.sum(np.exp(a))
```

```python
c = np.max(a)
np.exp(a - c) / np.sum(np.exp(a - c))
```

```python
# 重新实现softmax
def softmax(a):
    c = np.max(a)
    return np.exp(a - c) / np.sum(np.exp(a - c))
```

# softmax函数的特征

```python
a = np.array([0.3, 2.9, 4.0])
y = softmax(a)
print(y)
np.sum(y)
```
