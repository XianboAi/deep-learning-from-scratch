# 阶跃函数的实现

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
import numpy as np

def step_function(x):
    y = x > 0
    return y.astype(np.int_)
```

```python
step_function(np.array([-1.0, 1.0, 2.0]))
```

# 阶跃函数的图形

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x): # 第3种实现方式
    return np.array(x > 0, dtype=np.int_)

x = np.arange(-5, 5, 0.1)
y = step_function(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1) # 指定y轴范围 对于本图好像没什么区别
plt.show()
```

# sigmoid函数的实现

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.show()
```

# ReLU函数

```python
def relu(x):
    return np.maximum(0, x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.show()
```
