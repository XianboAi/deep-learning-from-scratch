# Affine层的实现

```python
class Affine:
    def __init__(self, W, b) -> None:  # 需要参数 所以需要保存 W 和 b 的梯度
        self.W = W
        self.b = b
        self.x = None
        self.dW = None
        self.db = None


    def forward(self, x):
        self.x = x
        out = np.dot(x, self.W) + self.b

        return out
    
    def backward(self, dout):
        self.db = np.sum(dout, axis=0)  # 为什么是axis=0？为什么是sum？ 因为就是竖着加的，每一行 + 一遍b
        self.dW = np.dot(self.x.T, dout)  # 这里为什么是这样画个图就知道了（见草稿
        dx = np.dot(dout, self.W.T)

        return dx
```

# Softmax-with-Loss层

```python
def softmax(x):
    if np.ndim(x) == 1:
        x = x.reshape(x.shape[0], -1)

    x_max = np.max(x, axis=1, keepdims=True)
    x_exp = np.exp(x - x_max)
    x_exp_sum = np.sum(x_exp, axis=1, keepdims=True)
    y = x_exp / x_exp_sum
    return y

def cross_entropy_error(y, t):
    if np.ndim(y) == 1:
        y = y.reshape(y.shape[0], -1)  # 这里第一个参数是 1！（这里就不改了）
        # t = t.reshape(t.shape[0], -1)  # 这里 t 不用转  t 本来就是1维  别犯病嗷
    
    loss = np.sum(y[np.arange(y.shape[0]), [t]])

class SoftnaxWithLoss():
    def __init__(self) -> None:
        self.y = None
        self.t = None
        self.loss = None
        self.dx = None

    def farward(self, x, t):
        y = softmax(x)
        self.y = y
        self.t = t
        self.loss = cross_entropy_error(y, t)

        return self.loss  #最后一层的输出是损失 loss
    
    def backward(self, dout=1):
        batch_size = self.t.shape[0]
        self.dx = (self.t - self.y) / batch_size

        return self.dx
```
