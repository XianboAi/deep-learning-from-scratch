# 激活函数层的实现

```python
# Relu层
class Relu:
    def  __init__(self) -> None:
        self.mask = None
    
    def forward(self, x):
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0

        return out

    def backword(self, dout):
        dout[self.mask] = 0
        dx = dout

        return dx
```

```python
# Sigmoid层
class Sigmoid():
    def __init__(self) -> None:  # 不需要更新参数 因此无需保存梯度 直接 return 即可
        self.out = None
    
    def forward(self, x):
        out = 1 / (1 + np.exp(-x))
        self.out = out

        return out
    
    def backward(self, dout):
        dx = dout * self.out * (1 - self.out)

        return dx
```
